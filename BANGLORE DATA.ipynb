```python
def func5(data):
    df_new = pd.DataFrame()
    for key,dt_fr in data.groupby("location"):
        m = np.mean(dt_fr["price_per_sqft"])
        s = np.std(dt_fr["price_per_sqft"])
        first_std_df = dt_fr[(dt_fr["price_per_sqft"]>=(m-s))&(dt_fr["price_per_sqft"]<=(m+s))]
        df_new = pd.concat([df_new,first_std_df],ignore_index=True)
    return df_new


    
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True,
    normalize=False,
    copy_X=True,
    n_jobs=None)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,mean_absolute_error,mean_squared_error
```

```python
model.fit(X_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
prediction=model.predict(X_test)
```

```python
print(mean_absolute_error(y_test,prediction))
```

```python
def predict_the_price(size,total_sqft,num_batrm,balcony,location,area_type):
    input = np.zeros(len(X.columns))
    input[0] = size
    input[1] = total_sqft
    input[2] = num_batrm
    input[3] = balcony

    id_loc = np.where(X.columns == location)[0][0]
    id_ar_ty = np.where(X.columns == area_type)[0][0]
    
    input[id_loc] = 1
    input[id_ar_ty] = 1

    return model.predict([input])

predict_the_price(3,1680,3,2,"other_loc","plot area")
```
